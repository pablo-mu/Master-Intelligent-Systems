```python
# Define the start and end poses.
T0 = SE3.Trans([1, 0.2, 0]) * SE3.RPY(0, 0, 3); #frame A
T1 = SE3.Trans([-1.2, -0.8, 0.5]) * SE3.RPY(3*pi/2, 3*pi/2, 0); #frame B
```

T0 = SE3.Trans([0.4, 0.2, 0]) * SE3.RPY(0, 0, 3);
T1 = SE3.Trans([-0.4, -0.2, 0.3]) * SE3.RPY(-pi/4, pi/4, -pi/2);

```python
#Iterpolate between the two poses	
Ts = T0.interp(T1,100)
```

```python
#Plot the movement of the frame towards the location of the frame B.
plotvol3(new = True, labels = True)
T0.plot(color = 'red', label = 'T0', frame = 'A')
T1.plot(color = 'blue', label = 'T1', frame = 'B')
HTML(Ts.animate(movie=True, dim=1.4, color = 'black'))
```
